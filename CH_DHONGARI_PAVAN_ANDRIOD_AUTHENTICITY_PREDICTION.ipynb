# Using DataWig to impute categorical column related apps #

```python
# Splitting original data set for imputing values for related apps
df_train, df_test = datawig.utils.random_split(original_data)
```

```python
# inputting required for algorithm
imputer = datawig.SimpleImputer(
    input_columns=['Category','Related apps'], # column(s) containing information about the column we want to impute
    output_column= 'Related apps', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )
```

```python
# applying algorithm
imputer.fit(train_df=df_train, num_epochs=50)
```

```python
related_apps = imputer.predict(relapps_data)
```

```python
# Checking metrics for validation
imputer.load_metrics()
```

```python
related_apps['Related apps'] = related_apps['Related apps'].fillna(related_apps['Related apps_imputed'])
```

```python
related_apps.drop(['Related apps_imputed_proba','Related apps_imputed'],axis=1,inplace=True)
```

```python
related_apps.head()
```

```python
# list of indexes of rows for which related apps replaced with some value
list_index = list(related_apps.index.values)
```

```python
# Dropping indexes from original dataset
original_data.drop(index=list_index,axis=0,inplace=True)
```

```python
# to convert number of ratings into sub groups
def rat_to_groups(no_of_ratings):
  '''
  function to convert ratings to resp translation
  '''
  if no_of_ratings <= 10:
    return 'ratings between 0-10'
  elif (no_of_ratings >= 10) and (no_of_ratings <1000 ):
    return 'ratings between 10 and 1000'
  else:
    return 'ratings greater than 1000'
```

```python
original_data['Rating_cate_dummy'] = original_data['Number of ratings'].apply(lambda x: rat_to_groups(x) )
plt.figure(figsize=(15,9))
ax = sns.countplot(x="Rating_cate_dummy", data= original_data, palette='husl')
plt.xlabel('Sub groups ')
plt.ylabel('Number of applications')
plt.title('Number of users ratings each app')
```

```python
# to convert prices to groups
def price_to_groups(no_of_cats):
  '''
  function to convert ratings to resp translation
  '''
  if no_of_cats == 0:
    return 'Free apps(price = 0)'
  elif (no_of_cats > 0) and (no_of_cats <=10 ):
    return 'app price between >0 and 10'
  else:
    return 'app price greater than 10'
```
